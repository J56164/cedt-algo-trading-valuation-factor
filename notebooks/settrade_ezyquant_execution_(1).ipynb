# <a id='toc3_'></a>[Connect Settrade](#toc0_)

- settrade open API: https://developer.settrade.com/open-api/manage-api-user
- ezyquant-execution: https://github.com/ezyquant/ezyquant-execution

```python
import os

from dotenv import load_dotenv

load_dotenv(override=True)

SETTRADE_APP_ID = os.environ["SETTRADE_APP_ID"]
SETTRADE_APP_SECRET = os.environ["SETTRADE_APP_SECRET"]
SETTRADE_APP_CODE = os.environ["SETTRADE_APP_CODE"]
SETTRADE_BROKER_ID = os.environ["SETTRADE_BROKER_ID"]
SETTRADE_PIN = os.environ["SETTRADE_PIN"]

SETTRADE_ACCOUNT_NO = os.environ["SETTRADE_ACCOUNT_NO"]
```

```python
settrade_equity_user = Investor(
    app_id=SETTRADE_APP_ID,
    app_secret=SETTRADE_APP_SECRET,
    app_code=SETTRADE_APP_CODE,
    broker_id=SETTRADE_BROKER_ID,
)
```

### <a id='toc3_1_1_'></a>[Execute Context](#toc0_)
- Get Data

```python
equity_ctx = ExecuteContext(settrade_user=settrade_equity_user, account_no=SETTRADE_ACCOUNT_NO, pin=SETTRADE_PIN)
```

```python
equity_ctx.port_value
```

```python
equity_ctx.get_account_info()
```

```python
equity_ctx.get_account_info().line_available
```

```python
pprint(equity_ctx.get_portfolios())
```

```python
pprint(equity_ctx.get_orders())
```

```python
pprint(equity_ctx.get_account_info())
```

### <a id='toc3_1_2_'></a>[Execute Context Symbol](#toc0_)

- Place order

```python
symbol = "PTT"
equity_ctx_symbol = ExecuteContextSymbol(settrade_user=settrade_equity_user,
                    account_no=SETTRADE_ACCOUNT_NO,
                    symbol=symbol,
                    pin=SETTRADE_PIN
                    )
```

```python
equity_ctx_symbol.place_order(
    side="Buy",
    volume=100,
    price=equity_ctx_symbol.best_bid_price,
    price_type="Limit",
    validity_type="Day"
)
```

```python
equity_ctx_symbol.place_order(
    side="Sell",
    volume=100,
    price=equity_ctx_symbol.best_ask_price,
    price_type="Limit",
    validity_type="Day"
)
```
